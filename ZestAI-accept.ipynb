# Dealing with irrevalent data & null values

```python
#Dropping columns with null values >= 60%
perc = 60.0 
min_count =  int(((100-perc)/100)*EDA_data.shape[0] + 1)
EDA_data = EDA_data.dropna( axis=1, thresh=min_count)
print("The number of rows & columns after dropping the data having 60% or more null values:", EDA_data.shape)   
```

```python
import numpy as np
df = pd.DataFrame()

df['Diff1'] = np.where( EDA_data['loan_amnt'] == EDA_data['funded_amnt'] , '1', '0')
df['Diff2'] = np.where( EDA_data['funded_amnt'] == EDA_data['funded_amnt_inv'] , '1', '0')
df['Diff3'] = np.where( EDA_data['loan_amnt'] == EDA_data['funded_amnt_inv'] , '1', '0')
df['Diff4'] = np.where( EDA_data['total_pymnt'] == EDA_data['total_pymnt_inv'] , '1', '0')
df['Diff5'] = np.where( EDA_data['total_rec_late_fee'] == EDA_data['recoveries'] , '1', '0')
df['Diff6'] = np.where( EDA_data['total_rec_late_fee'] == EDA_data['collection_recovery_fee'] , '1', '0')

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

fig, axes = plt.subplots(figsize=(20,15),nrows=1, ncols=4)
df['Diff1'].value_counts().plot(ax=axes[0], kind='bar')
df['Diff2'].value_counts().plot(ax=axes[1], kind='bar')
df['Diff3'].value_counts().plot(ax=axes[2], kind='bar')


fig, axes = plt.subplots(figsize=(20,15),nrows=1, ncols=3)
df['Diff4'].value_counts().plot(ax=axes[0], kind='bar')
df['Diff5'].value_counts().plot(ax=axes[1], kind='bar');
df['Diff6'].value_counts().plot(ax=axes[2], kind='bar');


plt.show()
```

```python
irrevalent_cols = ['id','sub_grade','grade','acc_now_delinq','policy_code','pymnt_plan','url','title','collection_recovery_fee',
                   'recoveries','issue_d','zip_code','fico_range_high','fico_range_low','out_prncp_inv','total_pymnt_inv',
                   'funded_amnt','funded_amnt_inv','emp_title','last_credit_pull_d','earliest_cr_line','initial_list_status',
                   'next_pymnt_d','collections_12_mths_ex_med','chargeoff_within_12_mths','num_tl_120dpd_2m','num_tl_30dpd',
                   'num_tl_90g_dpd_24m','hardship_flag','disbursement_method','debt_settlement_flag','addr_state','last_pymnt_d']
EDA_data.drop(columns =irrevalent_cols, inplace = True )
print("The number of rows & columns after dropping the null values:", EDA_data.shape)    
```

```python
# Plotting a Histogram
from matplotlib import pyplot as plt
EDA_data.loan_status.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Loan Status")
plt.ylabel('Number of people in each Loan status')
plt.xlabel('Loan Status');
```

```python
#Filtering data with Loan status "Fully Paid" & "Charged Off"
EDA_data1 = EDA_data[(EDA_data.loan_status == 'Fully Paid') | (EDA_data.loan_status == 'Charged Off')]
loan_stat = EDA_data1['loan_status'].value_counts()
print("We will consider only two types of loan_status:\n",loan_stat )
print("The number of rows & columns after filtering the values:", EDA_data1.shape)    
```

```python
left_null = EDA_data1.isnull().sum()
left_null.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Left Null values")
plt.xlabel('Number of null values left');
```

```python
#Filling the null values
EDA_data1.fillna(value=0, inplace=True)  
```

# Categorical Variables

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

fig, axes = plt.subplots(figsize=(20,15),nrows=1, ncols=4)
EDA_data1['term'].value_counts().plot(ax=axes[0], kind='bar')
EDA_data1['emp_length'].value_counts().plot(ax=axes[1], kind='bar')
EDA_data1['home_ownership'].value_counts().plot(ax=axes[2], kind='bar')
EDA_data1['application_type'].value_counts().plot(ax=axes[3], kind='bar');


fig, axes = plt.subplots(figsize=(20,15),nrows=1, ncols=3)
EDA_data1['verification_status'].value_counts().plot(ax=axes[0], kind='bar')
EDA_data1['loan_status'].value_counts().plot(ax=axes[1], kind='bar')
EDA_data1['purpose'].value_counts().plot(ax=axes[2], kind='bar');


plt.show()
```

```python
#Encoding Categorial variables
EDA_data1.term.replace((' 36 months', ' 60 months'),(1 ,0), inplace = True) 
EDA_data1.emp_length.replace(('10+ years','2 years','< 1 year','3 years','1 year','5 years','4 years','6 years','8 years','7 years','9 years'),(10,2,0.5,3,1,5,4,6,8,7,9), inplace  = True) 
EDA_data1.loan_status.replace(('Fully Paid', 'Charged Off'),(1,0), inplace = True) 
EDA_data1.application_type.replace(('Individual','Joint App'),(1,0), inplace  = True)
Categorical_var = ['home_ownership', 'verification_status', 'purpose']


for column in Categorical_var:
    tempdf = pd.get_dummies(EDA_data1[column], prefix=column)
    EDA_data1 = pd.merge(left=EDA_data1,right=tempdf,left_index=True,right_index=True,)
    EDA_data1 = EDA_data1.drop(columns=column)
```

# Interpretation

1. The lending club data set consist of features having a lot of null values.
2. The dataset also consists of many repetative features like loan_amnt & funded_amnt, total_pymnt & total_pymnt_inv etc.
3. There are many irrevalent features that are also present in the dataset.
4. After dealing with the null, repetative & irrevalent values the final shape of the dataset became from (2260701, 151) to (1345310, 76).
5. The feature "loan_status" is the target variable, it is something that we can predict from the study of this dataset.
5. For the feature "loan_status" out of the three most present values "Fully Paid", "Charged Off" & "Current", I am not considering the "Current" status as I don't know whether it will be paid or charged off in the future, so cannot be used for analysis & prediction. 
6. For all the object type Categorical variables, I have encoded them in the numeric form for the ease of model development. 
7. Finally, the correlation matrix roughly states the weightage of different features. Each of these features can be included in the model & regularization can be used to determine the final features in the final model.
